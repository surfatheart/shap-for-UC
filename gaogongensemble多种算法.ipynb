```python
import pandas as pd
import numpy as np
from sklearn.impute import KNNImputer

# 读取您的数据
file_path = r'D:\temp\fuke1130.xlsx'  # 替换成您的实际文件路径和文件名
df = pd.read_excel(file_path)

# 将 'null' 替换为 Pandas 的 NaN
df.replace('null', np.nan, inplace=True)

# 选择您关心的列，假设 'A-AU' 和 'x' 列是您关心的列
selected_columns = ['A-AU', 'x']

# 创建一个 DataFrame，只包含关心的列
selected_data = df[selected_columns]

# 创建 KNN 填充器，选择合适的邻居数量
knn_imputer = KNNImputer(n_neighbors=2)  # 可以调整邻居的数量

# 使用 KNN 填充缺失值
filled_data = knn_imputer.fit_transform(selected_data)

# 更新原始 DataFrame
df[selected_columns] = filled_data

# 显示更新后的 DataFrame
print(df)
```
